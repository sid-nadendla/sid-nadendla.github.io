# Searching in an array

Array Generator

```python
def generate_random_input(size):
    arr = np.random.randint(0, 20000, size=size)
    return arr
```

Linear Search

Linear-Search.png

Image source: [https://www.geeksforgeeks.org/linear-search/](https://www.geeksforgeeks.org/linear-search/)

```python
# Linear search function
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

Binary Search

image.png

Image Source: [https://www.geeksforgeeks.org/binary-search/](https://www.geeksforgeeks.org/binary-search/)

```python
def binary_search(arr, target):
    arr.sort()
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        guess = arr[mid]
        if guess == target:
            return mid
        if guess > target:
            high = mid - 1
        else:
            low = mid + 1
    return -1
```

Running Time

```python
# Function to measure binary search time
def measure_binary_search_time(size, target):
    arr = generate_random_input(size)
    start_time = time.time()
    binary_search(arr, target)
    return time.time() - start_time

# Define the start, end, and increments for the array sizes
start_array = 1000
end_array = 100000
increments = 1000

# Measure times for average and worst-case scenarios
avg_time_taken = []
worst_time_taken = []
for size in range(start_array, end_array+1, increments):
    # Repeat the measurement multiple times and calculate the average
    times = [measure_binary_search_time(size, -1) for _ in range(10)]
    avg_time_taken.append(np.mean(times))
    # For worst-case scenario, the target is outside of the array range
    worst_time_taken.append(measure_binary_search_time(size, size))

# Data Coordinates
x = np.arange(start_array, end_array+1, increments)
y = np.array(avg_time_taken)
z = np.array(worst_time_taken)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b', label='Avg Run Time')
plt.plot(x, z, 'r', label='Worst Run Time')

# Add Title
plt.title("Binary Search Run Times")

# Add Axes Labels
plt.xlabel("Size of Array")
plt.ylabel("Time Taken (seconds)")

# Display the legend and the plot
plt.legend()
plt.grid(True)
plt.show()
```

```python
def standard_matrix_multiply(A, B):
    n = len(A)
    result = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result
print("Standard Matrix Multiplication of A and B:\n", standard_matrix_multiply(A, B))
```

Strassen's Matrix Multiplication Algorithm Implementation

```python
import numpy as np

def strassen(A, B):
    """
    Strassen's Matrix Multiplication
    A and B should be numpy arrays.
    """
    n = A.shape[0]

    if n == 1:
        return A * B
    elif n <= 64:  # Base case threshold, can be adjusted
        return np.dot(A, B)

    # Initialize matrices
    new_size = n // 2
    A11, A12, A21, A22 = A[:new_size, :new_size], A[:new_size, new_size:], A[new_size:, :new_size], A[new_size:, new_size:]
    B11, B12, B21, B22 = B[:new_size, :new_size], B[:new_size, new_size:], B[new_size:, :new_size], B[new_size:, new_size:]

    # Compute the 7 products, recursively
    P1 = strassen(A11 + A22, B11 + B22)
    P2 = strassen(A21 + A22, B11)
    P3 = strassen(A11, B12 - B22)
    P4 = strassen(A22, B21 - B11)
    P5 = strassen(A11 + A12, B22)
    P6 = strassen(A21 - A11, B11 + B12)
    P7 = strassen(A12 - A22, B21 + B22)

    # Construct the result matrix from the computed parts
    C11 = P1 + P4 - P5 + P7
    C12 = P3 + P5
    C21 = P2 + P4
    C22 = P1 - P2 + P3 + P6

    C = np.vstack((np.hstack((C11, C12)), np.hstack((C21, C22))))
    return C
```
